# Crop Counting and Detection Using Transfer Learning based on Yolo

in this example we will use Yolo V8 for counting and detecting crops at emergence time. this is extremely important for breeding programs where we use AI to gain time and accuracy instead of labourious manuel traits estimation which is very expensive and error prone. 
the following steps are required 
 
 1. Create a virtual environment
 2. install labelme package for annotation (https://github.com/wkentaro/labelme)
 3. install ultralytics package for easy use of yolo (https://github.com/ultralytics/ultralytics)
 4. take video footage and generate sample images for annotation with labelme 
 5. Train Yolo on your custom dataset
 6. extract the weights  and run Inferecence

 all the dataset are available on my github account. This is an example running with less than 100 images  using yolo8nano 
 



```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
%ls
```

```python
model = YOLO('weights/crop-detection-countWeights.pt')  # Load the YOLOv8 model
```

## Inference on drone images


```python
image_path = "images/cowpea.jpg"
model.predict(source=image_path, show=False, save=True)
```

```python
img=cv2.imread("runs/detect/predict/cowpea.jpg")
plt.figure(figsize=(8,8))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
